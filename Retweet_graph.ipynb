# Personnes retweet√©es par des auteurs

```python
from lifranumTwitter import api
from collections import Counter
import csv
from datetime import datetime

now = datetime.now()
today = now.strftime("%Y%m%d")
filename = today + "_graph_retweet.csv"
print(filename)
```

```python
with open("AUTHORS_RESULTS/20201209_auteurs_base.txt", "r", encoding="utf-8") as file:
    authors = file.readlines()
    authors = [a.strip() for a in authors]
```

```python
import re

def get_retweeters(authors):
    retweeters = list()
    for author in authors : 
        print(author)
        results = api.GetUserTimeline(screen_name="@"+author, count=200)
        for result in results : 
            if re.search("^RT",result.full_text) : 
                if result.retweeted_status == None:
                    print(result)
                else : 
                    retweeter = result.retweeted_status.user.screen_name
                    retweeters.append((author, retweeter))
        compte = Counter(retweeters)
    return(compte)
retweeters = get_retweeters(authors)
```

```python
with open("AUTHORS_RESULTS/" + filename, "w")as graph_file:
    graph_writer = csv.writer(graph_file, delimiter='\t', quotechar='|', quoting=csv.QUOTE_MINIMAL)
    graph_writer.writerow(["Source", "Target", "Weight"])    
    for r in retweeters :     
        graph_writer.writerow([r[0], r[1], retweeters[r]])    
```

```python
import networkx as nx

graph = nx.read_edgelist(filename, delimiter="\t", data=[('Weight', float)])
nodes = nx.nodes(graph)
dict_nodes = dict()
for node in nodes : 
    if node in authors :
        dict_nodes[node] = "True"
    else :
        dict_nodes[node] = "False"
for node in nx.nodes(graph):
    if node in authors:
        nx.set_node_attributes(graph, dict_nodes,"authority")
        print(node)
        
```

```python
with open("AUTHORS_RESULTS/" + today + "nodes.csv", "w", encoding="utf-8") as nodes_file:
    graph_writer = csv.writer(nodes_file, delimiter='\t', quotechar='|', quoting=csv.QUOTE_MINIMAL)
    graph_writer.writerow(["Id", "Authority"])

    for node in dict_nodes :     
        graph_writer.writerow([node, dict_nodes[node]])
    
```
