```python
from lifranumTwitter import api, today
import csv 
authors = {}
filename = today + "_wattpad_authors.csv"
```

```python
# Chercher les personnes qui tweetent avec le #wattpad (durant les 7 derniers jours) 

results = api.GetSearch(term=f"#wattpad", lang="fr", count="1000")
for r in results:
    authors[r.user.screen_name] = [["hashtag"], r.user.description, "https://twitter.com/"+str(r.user.screen_name)]
print(authors)
```

```python
# Chercher les personnes qui tweetent avec un lien vers www.wattpad.com (durant les 7 derniers jours)

results = api.GetSearch(term=f"www.wattpad.com", lang="fr", count="1000")
for r in results : 
    try : 
        authors[r.user.screen_name][0].append("URL")
    except : 
        authors[r.user.screen_name] = [["URL"], r.user.description, "https://twitter.com/"+str(r.user.screen_name)]

        
```

```python
# Enregistrer le csv

with open(filename, "w", encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file, delimiter='|',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)
    for author in authors : 
        writer.writerow(authors[author])
```

```python
import requests
new_csv = []
with open("20210203_wattpad_authors.csv", "r", encoding="utf-8") as csv_file:
    reader = csv.reader(csv_file, delimiter='|')
    for row in reader : 
        row = list(row)
        author = row[2].replace("https://twitter.com/", "")
        user = api.GetUser(screen_name = author)
        if user.url : 
            response = requests.head(user.url)
            if response.is_redirect:
                url = response.headers['Location']
        else : 
            url = None
        row.append(url)
        if "wattpad" in str(url) : 
            row.append("True")
        else : 
            row.append("False")
        new_csv.append(row)

        
with open(filename, "w", encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file, delimiter='|',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)
    for row in new_csv : 
        writer.writerow(row)

```
