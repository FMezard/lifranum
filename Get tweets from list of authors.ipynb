```python
from lifranumTwitter import api
from collections import Counter
import csv
from datetime import datetime

now = datetime.now()
today = now.strftime("%Y%m%d")
filename = today + "_tweets_from_authors.csv"
print(filename)
```

```python
with open("AUTHORS_RESULTS/20200111_selection_auteurs.txt", "r", encoding="utf-8") as file:
    authors = file.readlines()
    authors = [a.strip() for a in authors]
```

```python
import re

def get_tweets(authors):
    tweets = list()
    for author in authors[:3] : 
        print(author)
        try : 
            results = api.GetUserTimeline(screen_name="@"+author, count=200)
        except TwitterError:
            pass
        for result in results : 
            if not re.search("^RT",result.full_text) :
                urls = [x.expanded_url for x in result.urls]
                hashtags = [h.text for h in result.hashtags]
                tweets.append([result.user.screen_name, result.user.name, result.full_text, hashtags, urls])
    return(tweets)
tweets = get_tweets(authors)
```

```python
print(tweets)
```

```python
with open(filename, "w", encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file, delimiter='|',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)
    for tweet in tweets : 
        writer.writerow(tweet)
```
