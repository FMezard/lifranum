```python
import community as community_louvain
import networkx as nx
```

```python
graph_hashtags = nx.read_edgelist("20200812_hashtags_l1graph.csv", data=[('Weight', int)],delimiter="\t")
print(len(graph_hashtags.nodes))
partition = community_louvain.best_partition(graph_hashtags)
```

```python
# liste des keys qui ont les mêmes values
import json
groups = dict()
for community_number in partition.values():
    groups[community_number] = [k for k,v in partition.items() if v == community_number]
with open("communities1.json", "w", encoding="utf-8") as file:
    json.dump(groups, file)
```

```python
graph_hashtags = nx.read_edgelist("20200812_hashtags_l2graph.csv", data=[('Weight', int)],delimiter="\t")
print(len(graph_hashtags.nodes()))
nodes = [u for (u,v) in nx.degree(graph_hashtags, weight="Weight") if v > 5]
print(len(nodes))
sub_graph = nx.subgraph(graph_hashtags, nodes)
partition = community_louvain.best_partition(sub_graph)
```

```python
# liste des keys qui ont les mêmes values
import json
groups = dict()
for community_number in partition.values():
    groups[community_number] = [k for k,v in partition.items() if v == community_number]
with open("communities.json", "w", encoding="utf-8") as file:
    json.dump(groups, file)
```
